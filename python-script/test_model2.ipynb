```python




model = load_model("../Models/generator_model.h5")
```

```python


def generate_image(image, model):
    height, width, c = image.shape

    img = cv2.resize(image, (256,256))
    print(img.shape)

    img = (img.astype(np.float32)/255.0).reshape(-1,256,256,3)

    prediction = model(img, training=False)



    return prediction

    


```

```python




def show(org, pred):
    
  plt.subplot(1, 2, 1)
  plt.title("org")
    # getting the pixel values between [0, 1] to plot it.
  plt.imshow(org)

  plt.subplot(1, 2, 2)
  plt.title("pred")
    # getting the pixel values between [0, 1] to plot it.
  plt.imshow(pred)


  plt.axis('on')
  plt.show()


```
