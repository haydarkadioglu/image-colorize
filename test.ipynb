```python
from library import colorize, findsize
import cv2
import matplotlib.pyplot as plt
```

```python

img = cv2.imread("Test/5.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

clr = colorize(img)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Colorized")
plt.imshow(clr)
plt.axis('off')

plt.show()
```
